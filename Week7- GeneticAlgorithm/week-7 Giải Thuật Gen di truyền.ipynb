```python
import numpy as np
import matplotlib.pyplot as plt

DNA_SIZE = 12 # do dai DNA
POP_SIZE = 120 # kick thuoc phan tu xin ra
CROSS_RATE = 0.9 # xac xuat giao phoi (cac DNA  giao nhau)
MUTATION_RATE = 0.09 # Xac suat dot bien 
N_GENERATIONS = 280
X_BOUND = [1, 10] # su tang va giam bac buoc cua x

def F(x): return np.sum(x*x + 1, axis=1) # Tinh gia tri lon nhat
 
#Tim cac gia tri khac 0
def get_fitness(pred): return np.max(pred) - pred + 1e-3# khoang cach la loi=))

def select(pop, fitness): #chon tu nhien  va bat fitnesss
    #chon he nhi phan
    idx = np .random.choice(np.arange(POP_SIZE),size=POP_SIZE, replace=True, p =fitness/fitness.sum())
    return pop[idx]

def crossover(s1,s2):
    crossover_prob = np.random.rand((DNA_SIZE))
    crossover_mask = crossover_prob < CROSS_RATE
    #luu 1
    buffer =s1.copy()
    
    # giao nhau va sinh ra 1 phan tu con
    s1[crossover_mask] = s2[crossover_mask]
    return (s1,s2)

def mutate(child):
    mutate_vector = np.random.randint(low=X_BOUND[0], high=X_BOUND[1], size=DNA_SIZE)
    
    mutate_prob = np.random.rand((DNA_SIZE))
    mutate_mask = mutate_prob < MUTATION_RATE
    
    child[mutate_mask] = mutate_vector[mutate_mask]
    return child


```

```python
pop = np.random.randint(low=X_BOUND[0],size=(POP_SIZE,DNA_SIZE)) # Khoi tao cac DNA dc bat

for g in range(N_GENERATIONS):
    F_values = F(pop) # Tinh toan gia tri ham = trich xuat DNA
    
    fitness = get_fitness(F_values)
    
    if g%5 == 0:
        print("Most fitness:", pop[np.argmax(fitness),:])
        
    pop = select(pop,fitness)
    parent_pop = pop.copy()
    
    for i in range(POP_SIZE//2):
        k1 = np.random.randint(0,POP_SIZE, size=1)
        k2 = np.random.randint(0,POP_SIZE, size=1)
        
        s1 = parent_pop[k1].copy()[0]
        s2 = parent_pop[k1].copy()[0]
        
        s1,s2 = crossover(s1,s2)
        
        s1 = mutate(s1)
        s2 = mutate(s2)
        
        # parent is replaced ny its child
        pop[i][:] = s1
        pop[i+1][:] = s1
```
